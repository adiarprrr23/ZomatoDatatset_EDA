```python
df_country=pd.read_excel('Zomatodataset\Country-Code.xlsx')
df_country.head()
```

```python
df_Final=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
df_Final.head()
```

```python
df_Final.dtypes
```

```python
df_Final.columns
```

```python
Country_names=df_Final.Country.value_counts().index
```

```python
Country_counts=df_Final.Country.value_counts().values
Country_counts
```

```python
# Top 3 countries using Zomato
plt.pie(Country_counts[:3],labels=Country_names[:3],autopct='%1.2f%%')
```

```python
df_Final.columns
```

```python
ratings=df_Final.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## Observations

1. 4.5-4.9-->Excellent
2. 4.0-4.4-->Very Good
3. 3.5-3.9-->Good
4. 2.5-3.4-->Average
5. 1.8-2.4-->Poor
6. 0.0-->Not Rated

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',data=ratings,palette=['white','red','orange','yellow','green','green'])
```

```python
sns.countplot(data=ratings,x='Rating color',palette=['blue','red','orange','yellow','green','green'])
```
